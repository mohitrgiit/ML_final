```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers.advanced_activations import LeakyReLU
from keras.optimizers import SGD, Adadelta, Adagrad, Adam, RMSprop
from skimage.util import crop
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
#setup model
model = Sequential()
model.add(Convolution2D(64, 5, 5, activation='relu', input_shape=(3, 64, 64), init='normal'))
model.add(Convolution2D(64, 5, 5, activation='relu', init='normal'))
model.add(Convolution2D(32, 3, 3, activation='relu', init='normal'))
model.add(Convolution2D(32, 3, 3, activation='relu', init='normal'))
model.add(Convolution2D(3, 5, 5, init='normal'))
model.load_weights('3_layer_net.h5')
model.compile(loss='mse', optimizer=Adam(lr=1e-4), metrics=['accuracy'])


num=4
test = np.load("data_preprocess/fin_data/test_pic.npy")
plt.figure(1)
plt.imshow(test[num])
# io.imsave("blur.jpg", test[num])
trans_test = test.transpose(0,3,1,2)
output = model.predict(trans_test)
temp_out = output-(output>1.)
output = output-(temp_out*(output>1.))
temp_out = output+(output<0.)
output = temp_out-(temp_out*(output<0.))
visualize = output.transpose(0,2,3,1)
plt.figure(2)
plt.imshow(visualize[num])
io.imsave("result.png", visualize[num])
```

```python
img=io.imread("1FIle.jpg")
print(img.shape)
img = cv2.resize(img, tuple(np.flipud(np.array(img.shape[:2])*2)), interpolation=cv2.INTER_CUBIC)
img = cv2.GaussianBlur(img, (3, 3), 0)
print(img.shape)
shape = list(img.shape)
shape.insert(0,-1)
img = img.reshape(tuple(shape))
img = (img.transpose(0,3,1,2))/255
out = model.predict(img)
temp_out = out-(out>1.)
out = out-(temp_out*(out>1))
plt.imshow(out[0].transpose(1,2,0))
io.imsave("1FIle_sr.png",out[0].transpose(1,2,0))
```
